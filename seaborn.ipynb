```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=sns.load_dataset("tips")
```

```python
df.head(3)
```

```python
df.shape
```

```python
sns.scatterplot(x="total_bill",y="tip",data=df)
plt.show()
```

```python
sns.lineplot(x='total_bill',y='size',data=df)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',data=df)
plt.show()
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
plt.show()
```

```python
sns.countplot(x='day',data=df)
plt.show()
```

```python
sns.countplot(x='sex',data=df)
plt.show()
```

```python
male=df[df['sex'] == 'Male'].value_counts().sum()
female=df[df['sex'] == 'Female'].value_counts().sum()
male,female
```
